```python
import matplotlib.pyplot as plt
import numpy as np
```

# グラフェンの描画

```python
a = 1
```

```python
a1 = a * np.array([np.sqrt(3)/2, 1/2])
a2 = a * np.array([np.sqrt(3)/2, -1/2])
```

```python
plt.figure(figsize=(10,10))
fig = plt.subplot()
for i in range(10):
    for j in range(10):
        a_type_pos = i*a1 + j*a2
        plt.scatter(*a_type_pos, color='black')  
        b_type_pos = a_type_pos + np.array([a/np.sqrt(3), 0])
        plt.scatter(*b_type_pos, color='red')
plt.show()
```

```python
plt.figure(figsize=(10,10))
fig = plt.subplot()
for i in range(-10,10):
    for j in range(-10, 10):
        a_type_pos = i*a1 + j*a2
        b_type_pos = a_type_pos + np.array([a/np.sqrt(3), 0])
        
        #line from b to a
        fig.plot([b_type_pos[0], b_type_pos[0] - a/np.sqrt(3)], [b_type_pos[1], b_type_pos[1]], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] + a/np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] + a/2], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] + a/np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] - a/2], color='black')
        fig.set_xlim(-2, 2)
        fig.set_ylim(-2, 2)
        fig.set_aspect('equal')

plt.show()
```

# 逆格子グラフェンの描画

```python
b1 = 2*np.pi/a * np.array([1/np.sqrt(3), 1])
b2 = 2*np.pi/a * np.array([1/np.sqrt(3), -1])
```

```python
plt.figure(figsize=(10,10))
fig = plt.subplot()
for i in range(10):
    for j in range(10):
        a_type_pos = i*b1 + j*b2
        fig.scatter(*a_type_pos, color='black')  
        b_type_pos = a_type_pos + np.array([0, 2*np.pi/a*2/3])
        fig.scatter(*b_type_pos, color='red')
fig.set_aspect('equal')
plt.show()
```

```python
plt.figure(figsize=(10,10))
fig = plt.subplot()
for i in range(-10, 10):
    for j in range(-10, 10):
        a_type_pos = i*b1 + j*b2 + np.array([0, 2*np.pi*2/a/3])
        b_type_pos = a_type_pos + np.array([0, 2*np.pi*2/a/3])
        #line from b to a
        fig.plot([b_type_pos[0], b_type_pos[0]], [b_type_pos[1], b_type_pos[1] - 2*np.pi/a*2/3], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] + 2*np.pi/a*2/3*np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] + 2*np.pi/a*2/3/2], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] - 2*np.pi/a*2/3*np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] + 2*np.pi/a*2/3/2], color='black')
        fig.set_xlim(-5, 5)
        fig.set_ylim(-5, 5)
        fig.set_aspect('equal')

plt.show()
```

# カッティングラインの描画

```python
n = 4
m = 2
```

```python
L = a * np.sqrt(n**2 + m**2 + n*m)
dt = L/np.pi
dr = np.gcd(2*m+n, 2*n+m)
t1 = (2*m + n)/dr
t2 = -1*(2*n + m)/dr
N = 2*(n**2 + m**2 + n*m)/dr
T = t1*a1 + t2*a2
```

```python
K1  = (-1*t2*b1 + t1*b2)/N
K2 = (m*b1 - n*b2)/N
```

```python
cutlinestart = -1* np.pi / np.linalg.norm(T) * K2 / np.linalg.norm(K2)
cutlineend = np.pi / np.linalg.norm(T) * K2 / np.linalg.norm(K2)
```

```python
plt.figure(figsize=(10,10))
fig = plt.subplot()
for i in range(-10, 10):
    for j in range(-10, 10):
        a_type_pos = i*b1 + j*b2 + np.array([0, 2*np.pi*2/a/3])
        b_type_pos = a_type_pos + np.array([0, 2*np.pi*2/a/3])
        #line from b to a
        fig.plot([b_type_pos[0], b_type_pos[0]], [b_type_pos[1], b_type_pos[1] - 2*np.pi/a*2/3], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] + 2*np.pi/a*2/3*np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] + 2*np.pi/a*2/3/2], color='black')
        fig.plot([b_type_pos[0], b_type_pos[0] - 2*np.pi/a*2/3*np.sqrt(3)/2], [b_type_pos[1], b_type_pos[1] + 2*np.pi/a*2/3/2], color='black')
        for myu in range(10):
            cutlinestart = -1* np.pi / np.linalg.norm(T) * K2 / np.linalg.norm(K2) + myu * K1
            cutlineend = np.pi / np.linalg.norm(T) * K2 / np.linalg.norm(K2) + myu * K1
            fig.plot([cutlinestart[0], cutlineend[0]], [cutlinestart[1], cutlineend[1]], color='black')
        fig.set_xlim(-5, 5)
        fig.set_ylim(-5, 5)
        fig.set_aspect('equal')

plt.show()
```
