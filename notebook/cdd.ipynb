```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rpm = 1000

cpoise = 0.294
gperml = 0.6548
```

https://labchem-wako.fujifilm.com/jp/question/013375.html
1cP（または、cps）= 1mPa・s

1g/ml = 1kg/L = 1000kg/m^3

```python
pas = cpoise / 1000
kgperm3 = gperml * 1000
```

```python
nendo = pas
mitudo = kgperm3
dounendo = nendo / mitudo
kaitensokudo = rpm / 60 * 2 * np.pi
```

```python
def makatsu(h0, t, omega, dounendo):
    return h0 / np.sqrt(1 + ((4 * (omega**2) * (h0**2) * t) / (dounendo * 3)))
```

```python
h0 = 1000
#micro meter
```

```python
tlist = np.linspace(0.1, 100, 100000)

hlist = np.array([makatsu(h0/1000, t, kaitensokudo, dounendo)*1000000 for t in tlist])
plt.plot(tlist, hlist, label='init_thick'+str(h0)+'$\mu$m')

h0 = 1
hlist = np.array([makatsu(h0/1000, t, kaitensokudo, dounendo)*1000000 for t in tlist])
plt.plot(tlist, hlist, label='init_thick'+str(h0)+'$\mu$m')

h0 = 0.01
hlist = np.array([makatsu(h0/1000, t, kaitensokudo, dounendo)*1000000 for t in tlist])
plt.plot(tlist, hlist, label='init_thick'+str(h0)+'$\mu$m')

plt.xlabel('Time [s]')
plt.ylabel('thickness [nm]')
plt.title(f'Thickness of a liquid film {rpm}rpm')
plt.legend()
plt.show()
```
